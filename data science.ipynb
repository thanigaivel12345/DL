```python
import pandas as pd


data1 = {
    'CarName': ['Nissan versa', 'alfa-romero Quadrifoglio', 'alfa-romero giulia'],
    'price': [5499, 16500, 13495],
    'fueltype': ['gas', 'gas', 'gas']
}

df1 = pd.DataFrame(data1)
print("DataFrame 1:")
print(df1)
```

```python

data2 = {
    'CarName': ['alfa-romero stelvio', 'audi 100 ls', 'volvo 246'],
    'price': [16500, 13950, 22470],
    'fueltype': ['gas', 'gas', 'diesel']
}

df2 = pd.DataFrame(data2)
print("\nDataFrame 2:")
print(df2)
```

```python
pd.concat([df1, df2], ignore_index=True)

```

```python

data3 = {
    'CarName': ['Nissan versa', 'alfa-romero Quadrifoglio', 'alfa-romero giulia'],
    'year': [2015, 2016, 2017],
    'Manufacturer': ['Nissan', 'Alfa-Romeo', 'Alfa-Romeo']
}

df3= pd.DataFrame(data3)
print("\nDataFrame 3:")
print(df3)

concatenated_df_horizontal = pd.concat([df1, df3], axis=1)

print("\nConcatenated DataFrame (horizontal):")
print(concatenated_df_horizontal)
```
